```python
# Решение уравнения x·sin(x) - 1 = 0 методом простой итерации
# Находим наименьший по модулю положительный корень. Точность: 10^-5
# для символьного дифференцирования sympy
import math                     
import sympy as sp              
```

```python
# Исходная функция, корни которой ищем
def f(x):
    """F(x) = x·sin(x) - 1"""
    return x * math.sin(x) - 1
```

```python
# Проверка существования корня на отрезке [1, 2] (по графику видно, что корень ≈1.114)
x1, x2 = 1, 2
if f(x1) * f(x2) < 0:
    print('Корень существует на отрезке [1, 2] (функция меняет знак)')
else:
    print('Корень не существует на этом отрезке')
```

```python
# Приведение уравнения к виду x = φ(x) для метода итераций
# Исходное уравнение: x·sin(x) = 1  =>  x = 1 / sin(x)
# Таким образом, φ(x) = 1 / sin(x) на отрезке [1,2] sin(x) > 0, поэтому знаменатель не обращается в ноль.
def phi(x):
    """Итерационная функция φ(x) = 1 / sin(x)"""
    return 1 / math.sin(x)
```

```python
# Аналитическое вычисление производной φ'(x) с помощью sympy
# Это нужно для проверки условия сходимости |φ'(x)| < 1 на отрезке.
# объявляем символьную переменную
x_sym = sp.Symbol('x', real=True)
# символьное выражение φ(x)          
phi_expr = 1 / sp.sin(x_sym)
# производная               
phi_derivative_expr = sp.diff(phi_expr, x_sym)   
print("\nСимвольное выражение φ(x):", phi_expr)
print("Символьное выражение φ'(x):", phi_derivative_expr)
```

```python
# Численная реализация производной (для подстановки значений)
def dphi(x):
    """Численное значение производной φ'(x) = -cos(x) / sin²(x)"""
    return -math.cos(x) / (math.sin(x) ** 2)
```

```python
# Проверка условия сходимости |φ'(x)| < 1 на отрезке [1, 2]
# Вычисляем модуль производной на концах отрезка и берём максимум.
a, b = 1, 2
dphi_a = abs(dphi(a))
dphi_b = abs(dphi(b))
max_dphi = max(dphi_a, dphi_b)

print(f"\n|φ'({a})| = {dphi_a:.6f}")
print(f"|φ'({b})| = {dphi_b:.6f}")
print(f"Максимум |φ'(x)| на [{a}, {b}] = {max_dphi:.6f}")

if max_dphi < 1:
    print("Условие сходимости выполняется: |φ'(x)| < 1 на всём отрезке.")
else:
    print("Условие сходимости НЕ выполняется. Итерации могут расходиться.")
```

```python
# Выбор начального приближения
# Возьмём середину отрезка [1,2] как начальное x0.
x0 = (a + b) / 2      # x0 = 1.5
print(f"\nНачальное приближение: x0 = {x0}")
```

```python
# Метод простой итерации
# Последовательно вычисляем x_{n+1} = φ(x_n)
# Останавливаемся, когда |x_{n+1} - x_n| < eps (eps = 1e-5)
# требуемая точность
eps = 1e-5 
# предыдущее приближение                    
x_prev = x0 
# счётчик итераций                   
iter_count = 0                

print("\nИтерации:")
print(f"{'n':<4} {'x_n':<12} {'x_{n+1}':<12} {'|x_{n+1}-x_n|':<16}")

while True:
    # вычисляем следующее приближение
    x_next = phi(x_prev) 
    # разность между соседними итерациями              
    diff = abs(x_next - x_prev)        
    print(f"{iter_count:<4} {x_prev:<12.8f} {x_next:<12.8f} {diff:<16.8f}")
    # достигнута нужная точность
    if diff < eps:                     
        root = x_next
        break
        # подготовка к следующей итерации
    x_prev = x_next                    
    iter_count += 1

    # Защита от бесконечного цикла (если что-то пошло не так)
    if iter_count > 100:
        print("Превышено максимальное число итераций (100).")
        root = x_next
        break
```

```python
# Вывод результата
print(f"Найденный корень: x = {root:.8f}")
print(f"Количество итераций: {iter_count+1}")
print(f"Значение функции в корне: f({root:.8f}) = {f(root):.2e}")
```
