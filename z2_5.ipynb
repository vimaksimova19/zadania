```python
# Определяем функцию, корень которой ищем

def f(x):
    """Исходное уравнение: x^3 + 3x^2 - 24x + 1 = 0"""
    return x**3 + 3*x**2 - 24*x + 1
```

```python
# Задаём отрезок [a, b], на котором ищем корень
# левая граница
a = 0       
# правая граница
b = 1       
```

```python
# Точность, с которой нужно найти корень
eps = 1e-5  # 0.00001
# Проверяем, есть ли корень на отрезке 
# Если f(a) и f(b) одного знака, то корня может не быть
if f(a) * f(b) < 0:
    print("Корень существует на отрезке [", a, ";", b, "] (функция меняет знак).")
else:
    print("На отрезке [", a, ";", b, "] корня может не быть (знаки одинаковые).")
```

```python
# Метод хорд (секущих)
# Идея: через точки (x0, f(x0)) и (x1, f(x1)) проводим прямую (хорду),
# её пересечение с осью Ox даёт новое приближение x2.
# Формула: x2 = x1 - f(x1)*(x1 - x0)/(f(x1)-f(x0))

# Начальные приближения: берём концы отрезка
# первое приближение
x0 = a
# второе приближение      
x1 = b     

# Счётчик итераций
iter_count = 0

# Итерационный процесс
while True:
    # Вычисляем следующее приближение по формуле хорд
    x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))

    # Увеличиваем счётчик шагов
    iter_count += 1

    # Проверяем, достигнута ли нужная точность
    # Критерий: модуль разности между двумя последними приближениями меньше eps
    if abs(x2 - x1) < eps:
        break

    # Подготавливаемся к следующей итерации:
    # старый x1 становится x0, а новое x2 становится x1
    x0 = x1
    x1 = x2
```

```python
# Вывод результата

print("\nРезультат метода хорд:")
print(f"Корень уравнения: x = {x2:.8f}")
print(f"Количество итераций: {iter_count}")
print(f"Значение функции в корне: f({x2:.8f}) = {f(x2):.2e}")
```
