```python
#  Импорт необходимых библиотек
from math import sin, log10          
import numpy as np                   
import matplotlib.pyplot as plt      
import matplotlib.ticker as ticker   
```

```python
# Разбор функции : x * sin(x) - 1 = 0
# Представим её в виде левой и правой частей: левая часть: y = x * sin(x), правая часть: y = 1
# Корни исходного уравнения – это точки пересечения этих двух графиков.

# Определяем функции для первой задачи
def f1(x):
    """Исходная функция f(x) = x*sin(x) - 1."""
    return x * sin(x) - 1

def y1_left(x):
    """Левая часть уравнения: y = x*sin(x)"""
    return x * sin(x)

def y1_right(x):
    """Правая часть уравнения: y = 1 (константа)"""
    return 1
```

```python
# Создаём массив значений x на интервале, где ожидаем корни (например, от -5 до 5)
# 500 равномерных точек от -5 до 5
x_vals1 = np.linspace(-5, 5, 500)          
# значения левой части
y_left_vals = [y1_left(x) for x in x_vals1]    
# значения правой части (везде 1)
y_right_vals = [y1_right(x) for x in x_vals1]  
```

```python
# Строим график
# создаём фигуру с осями
fig1, ax1 = plt.subplots(figsize=(10, 6))   
ax1.plot(x_vals1, y_left_vals, color='green', linewidth=2, label='y = x·sin(x)')
ax1.plot(x_vals1, y_right_vals, color='blue', linewidth=2, label='y = 1')

# Настройка графика
# основные деления по x через 1
ax1.xaxis.set_major_locator(ticker.MultipleLocator(1)) 
# дополнительные деления через 0.1  
ax1.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
# горизонтальная линия y=0
ax1.axhline(0, color='black', linestyle='--', linewidth=0.8)
# сетка  
ax1.grid(True, which='both', alpha=0.3)                  
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('График функций для уравнения x·sin(x) = 1')
ax1.legend()
```

```python
# Находим интервалы, где исходная функция f1(x) меняет знак. Делаю это перебором с шагом 0.1 по x.
# список для хранения интервалов
intervals1 = []
# шаг поиска        
step = 0.1
# начальная точка             
x = -5                 
while x < 5:
    # Если произведение значений функции на концах маленького интервала отрицательно, значит на этом интервале есть корень.
    if f1(x) * f1(x + step) < 0:
        intervals1.append((x, x + step))
    x += step
```

```python
# Выводим найденные интервалы
print("Результаты уравнения: x·sin(x) = 1")
print("Интервалы, содержащие корни (по знаку функции f(x) = x·sin(x)-1):")
for i, (a, b) in enumerate(intervals1):
    print(f"  Интервал {i+1}: [{a:.3f}, {b:.3f}]")


# Показываем график (после того, как всё выведено)
plt.show()
```

```python
# Разбор функции: x² - log10(x) - 6 = 0
#    Представим её в виде: левая часть: y = log10(x), правая часть: y = x² - 6
#    Область определения: x > 0 (логарифм)

# Определяем функции
def f2(x):
    """Исходная функция f(x) = x² - log10(x) - 6."""
    return x**2 - log10(x) - 6

def y2_left(x):
    """Левая часть уравнения: y = log10(x)"""
    return log10(x)

def y2_right(x):
    """Правая часть уравнения: y = x² - 6"""
    return x**2 - 6
```

```python
# Создаём массив x от 0.1 до 5 (логарифм определён при x>0)
# 500 точек от 0.1 до 5
x_vals2 = np.linspace(0.1, 5, 500)
# значения левой части          
y2_left_vals = [y2_left(x) for x in x_vals2]
# значения правой части   
y2_right_vals = [y2_right(x) for x in x_vals2]
```

```python
# Строим график
fig2, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(x_vals2, y2_left_vals, color='green', linewidth=2, label='y = log₁₀(x)')
ax2.plot(x_vals2, y2_right_vals, color='blue', linewidth=2, label='y = x² - 6')

# Настройка осей и сетки
ax2.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax2.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax2.axhline(0, color='black', linestyle='--', linewidth=0.8)
ax2.grid(True, which='both', alpha=0.3)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('График функций для уравнения log₁₀(x) = x² - 6')
ax2.legend()
```

```python
# Находим интервалы, где f2(x) меняет знак (шаг 0.1, начиная с x=0.1)
intervals2 = []
step = 0.1
x = 0.1
while x < 5:
    if f2(x) * f2(x + step) < 0:
        intervals2.append((x, x + step))
    x += step
```

```python
# Выводим интервалы для второго уравнения
print("Результаты ур-ния: x² - log₁₀(x) - 6 = 0")
print("Интервалы, содержащие корни (по знаку функции f(x) = x² - log₁₀(x) - 6):")
for i, (a, b) in enumerate(intervals2):
    print(f"  Интервал {i+1}: [{a:.3f}, {b:.3f}]")

# Показываем график
plt.show()
```
